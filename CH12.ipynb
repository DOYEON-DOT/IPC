```python
from google.colab.patches import cv2_imshow
from google.colab import drive
import numpy as np
import cv2
import matplotlib.pyplot as plt
drive.mount('/content/gdrive')
```

```python
img = cv2.imread('/content/gdrive/My Drive/IPC/Images/home3.jpg')
print(img.shape)
gimg = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
#plt.imshow(gimg, cmap='gray')
cv2_imshow(img)
```

```python
img2 = img.copy()
```

```python
edges = cv2.Canny(gimg, 100, 200)
```

```python
cv2_imshow(edges)
```

```python
lines = cv2.HoughLinesP(edges, 1, np.pi/180, 220)
```

```python
for line in lines:
  x1, y1, x2, y2 = line[0]
  cv2.line(img2, (x1, y1), (x2, y2), (0, 0, 255), 2)

cv2_imshow(img2)
```
